```python
import pandas as pd

df = pd.read_csv("raw/Historical Product Demand.csv")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head(3)
```

```python
import os

# Create processed folder if it doesn't exist
os.makedirs("data/processed", exist_ok=True)

# ── STEP 1: Standardise column names ──────────────────────────────────────────
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(f"[1/5] Columns: {df.columns.tolist()}")

# ── STEP 2: Parse dates ───────────────────────────────────────────────────────
before = len(df)
df["date"] = pd.to_datetime(df["date"], infer_datetime_format=True, errors="coerce")
df = df.dropna(subset=["date"])
print(f"[2/5] Dates parsed — dropped {before - len(df):,} unparseable rows")
print(f"      Date range: {df['date'].min().date()} → {df['date'].max().date()}")

# ── STEP 3: Clean demand values ───────────────────────────────────────────────
before = len(df)
df["order_demand"] = (
    df["order_demand"]
    .astype(str)
    .str.replace(r"[\(\)]", "", regex=True)
    .str.strip()
)
df["order_demand"] = pd.to_numeric(df["order_demand"], errors="coerce")
df = df.dropna(subset=["order_demand"])
df = df[df["order_demand"] > 0]
print(f"[3/5] Demand cleaned — dropped {before - len(df):,} negative/invalid rows")
print(f"      Demand range: {df['order_demand'].min():,.0f} — {df['order_demand'].max():,.0f}")

# ── STEP 4: Aggregate to weekly ───────────────────────────────────────────────
df = df.set_index("date")
weekly = (
    df.groupby(["product_code", "warehouse", "product_category"])["order_demand"]
    .resample("W-MON")
    .sum()
    .reset_index()
)
print(f"[4/5] Aggregated to weekly — {len(weekly):,} rows")
print(f"      Unique products  : {weekly['product_code'].nunique():,}")
print(f"      Unique warehouses: {weekly['warehouse'].nunique():,}")

# ── STEP 5: Filter to top 15 products ────────────────────────────────────────
top_products = (
    weekly.groupby("product_code")["order_demand"]
    .sum()
    .nlargest(15)
    .index.tolist()
)
weekly_top = weekly[weekly["product_code"].isin(top_products)].copy()
print(f"[5/5] Filtered to top 15 products — {len(weekly_top):,} rows")

# ── SAVE ──────────────────────────────────────────────────────────────────────
weekly_top.to_csv("data/processed/demand_clean.csv", index=False)
print(f"\n✅ Saved → data/processed/demand_clean.csv")

# ── SUMMARY ───────────────────────────────────────────────────────────────────
print("\nTop 15 products by total demand:")
top_summary = (
    weekly.groupby("product_code")["order_demand"]
    .sum()
    .nlargest(15)
    .reset_index()
)
top_summary.columns = ["product_code", "total_demand"]
top_summary["total_demand"] = top_summary["total_demand"].apply(lambda x: f"{x:,.0f}")
print(top_summary.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

# Style
plt.rcParams["figure.dpi"] = 150
plt.rcParams["font.family"] = "sans-serif"
sns.set_style("whitegrid")

os.makedirs("charts", exist_ok=True)

# ── CHART 1: Total weekly demand over time (all top 15 products) ──────────────
fig, ax = plt.subplots(figsize=(12, 4))

total_weekly = weekly_top.groupby("date")["order_demand"].sum().reset_index()
ax.plot(total_weekly["date"], total_weekly["order_demand"], 
        color="#378ADD", linewidth=1.2)
ax.fill_between(total_weekly["date"], total_weekly["order_demand"], 
                alpha=0.1, color="#378ADD")

ax.set_title("Total weekly demand — top 15 products (2011–2017)", 
             fontsize=13, fontweight="500", pad=12)
ax.set_xlabel("")
ax.set_ylabel("Total order demand", fontsize=11)
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.xaxis.set_major_locator(mdates.YearLocator())
plt.tight_layout()
plt.savefig("charts/chart1_total_demand_over_time.png")
plt.show()
print("✅ Chart 1 saved")

# ── CHART 2: Total demand by product category ─────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 4))

cat_demand = (
    weekly_top.groupby("product_category")["order_demand"]
    .sum()
    .sort_values(ascending=True)
)
bars = ax.barh(cat_demand.index, cat_demand.values, color="#1D9E75", height=0.6)
ax.bar_label(bars, labels=[f"{v/1e6:.0f}M" for v in cat_demand.values], 
             padding=5, fontsize=10)
ax.set_title("Total demand by product category", 
             fontsize=13, fontweight="500", pad=12)
ax.set_xlabel("Total order demand", fontsize=11)
ax.set_xlim(0, cat_demand.max() * 1.2)
plt.tight_layout()
plt.savefig("charts/chart2_demand_by_category.png")
plt.show()
print("✅ Chart 2 saved")

# ── CHART 3: Top 15 products by total demand ──────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5))

prod_demand = (
    weekly_top.groupby("product_code")["order_demand"]
    .sum()
    .sort_values(ascending=True)
)
bars = ax.barh(prod_demand.index, prod_demand.values, color="#7F77DD", height=0.6)
ax.bar_label(bars, labels=[f"{v/1e6:.0f}M" for v in prod_demand.values], 
             padding=5, fontsize=9)
ax.set_title("Top 15 products by total demand (2011–2017)", 
             fontsize=13, fontweight="500", pad=12)
ax.set_xlabel("Total order demand", fontsize=11)
ax.set_xlim(0, prod_demand.max() * 1.2)
plt.tight_layout()
plt.savefig("charts/chart3_top15_products.png")
plt.show()
print("✅ Chart 3 saved")

# ── CHART 4: Demand over time for top product ─────────────────────────────────
top_product = "Product_1359"
fig, ax = plt.subplots(figsize=(12, 4))

prod_ts = weekly_top[weekly_top["product_code"] == top_product].copy()
ax.plot(prod_ts["date"], prod_ts["order_demand"], 
        color="#D85A30", linewidth=1.2)
ax.fill_between(prod_ts["date"], prod_ts["order_demand"], 
                alpha=0.1, color="#D85A30")
ax.set_title(f"Weekly demand — {top_product} (2011–2017)", 
             fontsize=13, fontweight="500", pad=12)
ax.set_xlabel("")
ax.set_ylabel("Order demand", fontsize=11)
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.xaxis.set_major_locator(mdates.YearLocator())
plt.tight_layout()
plt.savefig("charts/chart4_top_product_demand.png")
plt.show()
print("✅ Chart 4 saved")
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# ── CONFIG ────────────────────────────────────────────────────────────────────
TARGET_PRODUCT = "Product_1359"
HOLDOUT_WEEKS  = 12   # last 12 weeks used for evaluation

# ── PREPARE DATA ──────────────────────────────────────────────────────────────
prod_df = (
    weekly_top[weekly_top["product_code"] == TARGET_PRODUCT]
    .groupby("date")["order_demand"]
    .sum()
    .reset_index()
)
prod_df.columns = ["ds", "y"]
prod_df = prod_df.sort_values("ds").reset_index(drop=True)

print(f"Product      : {TARGET_PRODUCT}")
print(f"Total weeks  : {len(prod_df)}")
print(f"Train weeks  : {len(prod_df) - HOLDOUT_WEEKS}")
print(f"Holdout weeks: {HOLDOUT_WEEKS}")

# ── TRAIN / TEST SPLIT ────────────────────────────────────────────────────────
train = prod_df.iloc[:-HOLDOUT_WEEKS]
test  = prod_df.iloc[-HOLDOUT_WEEKS:]

# ── PROPHET MODEL ─────────────────────────────────────────────────────────────
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    changepoint_prior_scale=0.05,
    interval_width=0.95
)
model.fit(train)

# Predict over full range + holdout
future   = model.make_future_dataframe(periods=HOLDOUT_WEEKS, freq="W")
forecast = model.predict(future)

# ── EVALUATE ON HOLDOUT ───────────────────────────────────────────────────────
forecast_holdout = forecast.iloc[-HOLDOUT_WEEKS:][["ds", "yhat", "yhat_lower", "yhat_upper"]]
forecast_holdout = forecast_holdout.reset_index(drop=True)
test = test.reset_index(drop=True)

mae  = mean_absolute_error(test["y"], forecast_holdout["yhat"])
rmse = np.sqrt(mean_squared_error(test["y"], forecast_holdout["yhat"]))
mape = (abs(test["y"] - forecast_holdout["yhat"]) / test["y"].replace(0, np.nan)).mean() * 100

print(f"\nModel performance on holdout ({HOLDOUT_WEEKS} weeks):")
print(f"  MAE  : {mae:,.0f}")
print(f"  RMSE : {rmse:,.0f}")
print(f"  MAPE : {mape:.1f}%")

# ── CHART 5: Forecast vs Actual ───────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(13, 5))

ax.plot(train["ds"], train["y"], color="#378ADD", linewidth=1, label="Training data")
ax.plot(test["ds"],  test["y"],  color="#1D9E75", linewidth=1.5, label="Actual (holdout)")
ax.plot(forecast_holdout["ds"], forecast_holdout["yhat"], 
        color="#D85A30", linewidth=1.5, linestyle="--", label="Prophet forecast")
ax.fill_between(forecast_holdout["ds"],
                forecast_holdout["yhat_lower"],
                forecast_holdout["yhat_upper"],
                alpha=0.15, color="#D85A30", label="95% confidence interval")

ax.axvline(test["ds"].iloc[0], color="gray", linestyle=":", linewidth=1)
ax.text(test["ds"].iloc[0], ax.get_ylim()[1]*0.95, 
        " forecast start", fontsize=9, color="gray")

ax.set_title(f"Prophet forecast vs actual — {TARGET_PRODUCT} (12-week holdout)",
             fontsize=13, fontweight="500", pad=12)
ax.set_ylabel("Order demand", fontsize=11)
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig("charts/chart5_forecast_vs_actual.png")
plt.show()
print("✅ Chart 5 saved")
```

```python
# ── STEP 4: INVENTORY RECOMMENDATIONS ────────────────────────────────────────

print("=" * 55)
print("  Inventory Recommendations — Top 15 Products")
print("=" * 55)

results = []

for product in weekly_top["product_code"].unique():
    
    prod = (
        weekly_top[weekly_top["product_code"] == product]
        .groupby("date")["order_demand"]
        .sum()
        .reset_index()
    )
    prod.columns = ["ds", "y"]
    prod = prod.sort_values("ds").reset_index(drop=True)
    
    if len(prod) < 20:
        continue
    
    # Train Prophet
    m = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        changepoint_prior_scale=0.05
    )
    m.fit(prod)
    
    # Forecast next 4 weeks
    future   = m.make_future_dataframe(periods=4, freq="W")
    forecast = m.predict(future)
    next_4wk = forecast.tail(4)["yhat"].clip(lower=0).sum()
    
    # Inventory calculations
    avg_weekly   = prod["y"].mean()
    std_weekly   = prod["y"].std()
    lead_time    = 2      # weeks
    z            = 1.65   # 95% service level
    
    safety_stk   = round(z * std_weekly * np.sqrt(lead_time))
    reorder_pt   = round(avg_weekly * lead_time + safety_stk)
    
    results.append({
        "product_code"        : product,
        "avg_weekly_demand"   : round(avg_weekly),
        "forecast_next_4wks"  : round(next_4wk),
        "safety_stock"        : round(safety_stk),
        "reorder_point"       : round(reorder_pt),
        "volatility_cv"       : round(std_weekly / avg_weekly * 100, 1)
    })

recommendations = pd.DataFrame(results).sort_values(
    "avg_weekly_demand", ascending=False
).reset_index(drop=True)

# Save
recommendations.to_csv("data/processed/inventory_recommendations.csv", index=False)

# Print clean table
print("\n")
print(recommendations.to_string(index=False))
print(f"\n✅ Saved → data/processed/inventory_recommendations.csv")
```

```python
# ── CHART 6: Inventory recommendations visual ─────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: Safety stock vs avg demand
ax1 = axes[0]
x = range(len(recommendations))
width = 0.35

bars1 = ax1.bar([i - width/2 for i in x], 
                recommendations["avg_weekly_demand"],
                width, label="Avg weekly demand", color="#378ADD", alpha=0.85)
bars2 = ax1.bar([i + width/2 for i in x],
                recommendations["safety_stock"],
                width, label="Safety stock required", color="#D85A30", alpha=0.85)

ax1.set_xticks(list(x))
ax1.set_xticklabels(
    [p.replace("Product_", "P") for p in recommendations["product_code"]],
    rotation=45, ha="right", fontsize=8
)
ax1.set_title("Safety stock vs average weekly demand",
              fontsize=12, fontweight="500", pad=10)
ax1.set_ylabel("Units", fontsize=10)
ax1.legend(fontsize=9)

# Right: Volatility (CV) by product — highlights risky products
ax2 = axes[1]
colors = ["#E24B4A" if cv > 100 else "#EF9F27" if cv > 60 else "#1D9E75" 
          for cv in recommendations["volatility_cv"]]

bars = ax2.barh(
    [p.replace("Product_", "P") for p in recommendations["product_code"]],
    recommendations["volatility_cv"],
    color=colors, height=0.6
)
ax2.bar_label(bars, labels=[f"{v:.0f}%" for v in recommendations["volatility_cv"]],
              padding=4, fontsize=8)
ax2.axvline(100, color="#E24B4A", linestyle="--", linewidth=1, alpha=0.5)
ax2.axvline(60,  color="#EF9F27", linestyle="--", linewidth=1, alpha=0.5)
ax2.set_title("Demand volatility by product (CV %)\nRed >100% · Amber >60% · Green ≤60%",
              fontsize=12, fontweight="500", pad=10)
ax2.set_xlabel("Coefficient of variation (%)", fontsize=10)
ax2.set_xlim(0, recommendations["volatility_cv"].max() * 1.2)

plt.tight_layout()
plt.savefig("charts/chart6_inventory_recommendations.png", bbox_inches="tight")
plt.show()
print("✅ Chart 6 saved")

# ── FINAL SUMMARY ─────────────────────────────────────────────────────────────
print("\n" + "=" * 55)
print("  Analysis complete — summary")
print("=" * 55)
print(f"  Products analysed    : {len(recommendations)}")
print(f"  High-risk products   : {(recommendations['volatility_cv'] > 100).sum()} (CV > 100%)")
print(f"  Medium-risk products : {((recommendations['volatility_cv'] > 60) & (recommendations['volatility_cv'] <= 100)).sum()} (CV 60–100%)")
print(f"  Stable products      : {(recommendations['volatility_cv'] <= 60).sum()} (CV ≤ 60%)")
print(f"\n  Charts saved to      : charts/")
print(f"  Data saved to        : data/processed/")
```
